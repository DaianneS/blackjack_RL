# Ambiente Blackjack com Contagem de Cartas

## Objetivo
Este notebook documenta e valida a implementação do ambiente do jogo Blackjack que será utilizado nas etapas de simulação, comparação de políticas e treinamento do agente com Q-Learning.

O diferencial deste projeto é a modelagem de um ambiente realista de cassino, afastando-se da premissa de um baralho infinito. O foco desta etapa é validar:

- A representação do estado (que agora inclui a "temperatura" da mesa);
- A mecânica do jogo e o cálculo das recompensas.

## Regras Adotadas no Ambiente

Neste projeto, o Blackjack foi modelado com as seguintes regras de cassino:

- **O Baralho (Shoe):** O jogo utiliza um *shoe* composto por 6 baralhos (312 cartas no total). Quando restar menos de 20% das cartas, o montante é reembaralhado.
- **Valores das Cartas:** As cartas possíveis são de 1 a 10. As figuras (Valete, Dama, Rei) valem 10. O Ás (representado por 1) pode valer 1 ou 11, dependendo da situação mais vantajosa para não estourar 21.
- **Ações do Jogador:**
  - `hit`: pedir mais uma carta;
  - `stick`: parar.
- **Regra do Dealer:** O dealer é obrigado a pedir cartas até atingir uma soma de pelo menos 17 pontos.
- **Contagem de Cartas (Hi-Lo):** O ambiente rastreia automaticamente o *Running Count* e calcula o *True Count* (Running Count / baralhos restantes).
  - Cartas 2 a 6: +1
  - Cartas 7 a 9: 0
  - Cartas 10 e Ás: -1

## Testando o Baralho e a Contagem de Cartas (Hi-Lo)

Antes de simular uma partida inteira, vamos instanciar o nosso baralho (`ShoeBlackjack`) e comprar algumas cartas para verificar se o sistema de contagem está funcionando corretamente e atualizando o *True Count*.

```python
# Instanciando um shoe padrão de cassino (6 baralhos)
shoe_teste = ShoeBlackjack(num_baralhos=6)

print(f"Cartas iniciais no shoe: {len(shoe_teste.cartas)}")
print("-" * 40)

# Vamos simular a compra de 10 cartas e observar a contagem
for i in range(1, 11):
    carta = shoe_teste.comprar_carta()
    rc = shoe_teste.running_count
    tc = shoe_teste.get_true_count()
    print(f"Carta comprada: {carta:2} | Running Count: {rc:2} | True Count: {tc:2}")

print("-" * 40)
print(f"Cartas restantes no shoe: {len(shoe_teste.cartas)}")
```

A estrutura do ambiente está validada! O baralho se comporta de maneira finita, o sistema Hi-Lo está rastreando o histórico das cartas perfeitamente, e as partidas fluem extraindo cartas desse mesmo monte.

A inclusão do *True Count* no estado aumenta a complexidade do problema, mas fornece a informação matemática necessária para que o nosso futuro agente de Q-Learning aprenda não apenas as regras do Blackjack, mas também a nobre arte de contar cartas.
